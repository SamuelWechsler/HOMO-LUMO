```python
def Lambda(d, N, z):
    
    h = 6.262608015 * 10 ** (-34)
    m_e = 9.10938356 * 10 ** (-31)
    c = 2.99792458 * 10 ** 8
    d = d * 10 ** (-12)
    N_A = 6.02214076 * 10 ** (23)
    
    
    delta_E = N_A * (h**2 * (N+1)) /   (8 * m_e * (z+1)**2 * d**2 * 1000) 
    
    Lambdaa = 10 ** 9 * (8 * m_e * c * (z+1)**2 * d**2) / (h * (N+1))
    
    
    return Lambdaa
    
```

```python
def delta_E(d, N, z):
    
    h = 6.262608015 * 10 ** (-34)
    m_e = 9.10938356 * 10 ** (-31)
    c = 2.99792458 * 10 ** 8
    d = d * 10 ** (-12)
    N_A = 6.02214076 * 10 ** (23)
    
    
    delta_E = N_A * (h**2 * (N+1)) /   (8 * m_e * (z+1)**2 * d**2 * 1000) 
    
    Lambdaa = (8 * m_e * c * (z+1)**2 * d**2) / (h * (N+1)) * 10 ** 9
    
    return delta_E
```

```python
def delta_E_2(N):
    result = 1.88 * 10 * 3 * (N+1)**2 / N
    return result
```

```python
def lambda_2(N):
    result = 63.71 * N**2 / (N+1)
    return result
```
